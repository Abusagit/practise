```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 10)  # 10 чисел от 0 до 5
y = x ** 2  # квадрат от икс
print(x)  #
print(y)
```

```python
plt.figure()  # строим график
plt.plot(x, y, 'r')  # значения массивов, параметр - линия красным цветом
plt.xlabel('x')  # метка для оси Х - называется Х
plt.ylabel('y')
plt.title('title')  # Название графика
plt.show()  # ПОКАЖУ ЧТО ПОЛУЧИЛОСЬ
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # left, bottom, width, height (range 0 to 1)  - координаты на рисунке, 
                                            # так как надо еще оставить место
axes.plot(x, y, 'r')  # оси

# Подписи:
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # inset axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('My title')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title')
plt.show()
```

## 2 plots beside

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# в цикле указываем параметры каждого из графиков
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig.tight_layout()
```

### Set figures`s size

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('changed sized')
plt.show()
```

### Legend + 2 graphs on 1 plot

```python
fig, ax = plt.subplots()

ax.plot(x, x ** 2, 'r', label='y = x ** 2')
ax.plot(x, x ** 3, label='y = x ** 3')
ax.legend(loc=2)  # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('with legend')
plt.show()
```

### Histograms

```python
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(n)
axes[0].set_title('Default histogram')
axes[0].set_xlim(min(n), max(n))

axes[1].hist(n, cumulative=True, bins=50)  # Параметры - кумулятивная, число столбцов - 50
axes[1].set_title('Cumulative detailed histogram')
axes[1].set_xlim(min(n), max(n))

plt.show()
```
