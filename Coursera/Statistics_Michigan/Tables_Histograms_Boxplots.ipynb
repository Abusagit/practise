```python
import seaborn as sns
import matplotlib.pyplot as plt

# load in the data set
tips_data = sns.load_dataset('tips')
```

```python
print(tips_data.head())
```

# Describing data

```python
# Summary statistics
tips_data.describe()
```

## Histograms

```python
# Total bill
sns.distplot(tips_data['total_bill'], kde=False).set_title('Histogram of total bills')

plt.show()
# Unimodal, right-skewed
```

```python
sns.distplot(tips_data['tip']).set_title('Hist of tip')

plt.show()
```

### Both plots

```python
sns.distplot(tips_data['total_bill'], kde=False)
sns.distplot(tips_data['tip'], kde=False).set_title('Histogram of both')
plt.show()
```

## Creating boxplots

```python
sns.boxplot(tips_data['total_bill']).set_title('Boxplot of total bill')
plt.show()
```

```python
sns.boxplot(tips_data['tip']).set_title('Boxplot of tip')
plt.show()
```

```python
# +-+-+-+-
# Both of them
sns.boxplot(tips_data['total_bill'])
sns.boxplot(tips_data['tip']).set_title('Boxplot of total bill tip')
plt.show()
```

## Creating histograms and boxplots plotted by Groups

```python
# Plotting tips grouped by smoking
sns.boxplot(x=tips_data['tip'], y=tips_data['smoker'])

#    
g = sns.FacetGrid(tips_data, row='smoker')
g = g.map(plt.hist, 'tip')

plt.show()
```

```python
# Plotting tips grouped by time of the day
sns.boxplot(x = tips_data['tip'], y = tips_data['time'])
plt.show()

g = sns.FacetGrid(tips_data, row = 'time')
g = g.map(plt.hist, 'tip')
```

```python
# Plotting tips grouped by time of the day of the week
sns.boxplot(x = tips_data['tip'], y = tips_data['day']).set_title('Boxplots of tips by the day of the week')
plt.show()

s = sns.FacetGrid(tips_data, row = 'day')
s = s.map(plt.hist, 'tip')

# ==> Sunday have bigger amount of tips
```
