# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'Cartwheeldata.csv'

data_frame = pd.read_csv(url)

# Create Scatterplot:
sns.lmplot(x='Wingspan', y='CWDistance', data=data_frame)

plt.show()
```

```python
# Scatterplot arguments
sns.lmplot(x='Wingspan', y='CWDistance', data=data_frame,
           fit_reg=False, # No regression line)
           hue='Gender') # Color by evolution stage
plt.show()
```

```python
sns.lmplot(x='Wingspan', y='CWDistance', data=data_frame,
           hue='Gender') # Color by evolution stage
plt.show()
```

```python
# Construct Cartwheel distance plot
sns.swarmplot(x='Gender', y='CWDistance', data=data_frame)

plt.show()
```

## Boxplots


```python
sns.boxplot(data=data_frame.loc[:, ['Age', 'Height',  "Wingspan", 'CWDistance', 'Score']])

plt.show()
```

```python
# Male boxplot

sns.boxplot(data=data_frame.loc[data_frame['Gender'] == 'M', ['Age', 'Height',  "Wingspan", 'CWDistance', 'Score']])

plt.show()
```

```python
# Female boxplot
sns.boxplot(data=data_frame.loc[data_frame['Gender'] == 'F', ['Age', 'Height',  "Wingspan", 'CWDistance', 'Score']])

plt.show()
```

```python
# Male boxplot
sns.boxplot(data=data_frame.loc[data_frame['Gender'] == 'M', ['Score']])

plt.show()
```

```python
# Female

sns.boxplot(data=data_frame.loc[data_frame['Gender'] == 'F', ["Score"]])

plt.show()
```

## Histogram

```python
# Distribution Plot (a.k.a. Histogram)
a = sns.distplot(data_frame.CWDistance)

plt.show()
```

## Count plot

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Gender', data=data_frame)

plt.show()
```

```python
sns.countplot(x='Gender', data=data_frame)
plt.xticks(rotation=-45)

plt.show()

```
